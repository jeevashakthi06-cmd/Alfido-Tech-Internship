```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/kaggle/input/task-1/zomato.csv")  
df['rate_clean'] = df['rate'].astype(str).str.replace('/5', '', regex=False)
df['rate_clean'] = pd.to_numeric(df['rate_clean'], errors='coerce')
plt.figure(figsize=(6, 4))
plt.hist(df['rate_clean'].dropna(), bins=10)
plt.xlabel("Ratings")
plt.ylabel("Frequency")
plt.title("Distribution of Restaurant Ratings")
plt.tight_layout()
plt.show()
```
