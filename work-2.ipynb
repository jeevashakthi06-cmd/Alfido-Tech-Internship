```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/kaggle/input/loandataset/loan_prediction.csv")
plt.figure(figsize=(6,4))
df['Loan_Status'].value_counts().plot(kind='bar')
plt.title("Loan Status Count")
plt.xlabel("Loan Status")
plt.ylabel("Count")
plt.show()
plt.figure(figsize=(6,4))
plt.hist(df['ApplicantIncome'], bins=20)
plt.title("Distribution of Applicant Income")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()
plt.figure(figsize=(6,4))
df.boxplot(column='ApplicantIncome', by='Loan_Status')
plt.title("Applicant Income by Loan Status")
plt.suptitle("")  # removes extra title
plt.xlabel("Loan Status")
plt.ylabel("Income")
plt.show()
plt.figure(figsize=(6,4))
plt.scatter(df['ApplicantIncome'], df['LoanAmount'])
plt.title("Applicant Income vs Loan Amount")
plt.xlabel("Income")
plt.ylabel("Loan Amount")
plt.show()
plt.figure(figsize=(6,6))
df['Property_Area'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Property Area Distribution")
plt.ylabel("")
plt.show()
```
